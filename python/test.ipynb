```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
t0=-1
t1=3.31
q0=1.5
q1=1.93
v0=0
v1=0
#利用公式（1-22）求系数
h=q1-q0
T=t1-t0
a0=q0
a1=v0
a2=(3*h-(2*v0+v1)*T)/(T*T)
a3=(-2*h+(v0+v1)*T)/(T*T*T)
print(a0,a1,a2,a3)

t = np.linspace(t0,t1,20)
q = a0+a1*pow((t-t0),1)+a2*pow((t-t0),2)+a3*pow((t-t0),3)
q1 = a1+2*a2*pow((t-t0),1)+3*a3*pow((t-t0),2)
q2 = 2*a2+6*a3*pow((t-t0),1)
kappa = abs(q2) / pow((1 + q1*q1),3/2) 
```

```python
plt.plot(t,q,label="q")
plt.plot(t,q1,label="k")
plt.plot(t,q2,label="dk")
plt.grid("on")
plt.legend()
```
